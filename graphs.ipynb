```python
import matplotlib.pyplot as plt
import csv
```

```python
X = []
Y = []

plt.figure(figsize=(15, 9))
 
with open('simple_mult_one_thread.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=' ')
     
    for ROWS in plotting:
        X.append(int(ROWS[0]))
        Y.append(float(ROWS[1]))
 
plt.scatter(X, Y, label = 'simple', s=100, alpha=0.7)

X = []
Y = []
 
with open('block_mult_one_thread.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=' ')
     
    for ROWS in plotting:
        X.append(int(ROWS[0]))
        Y.append(float(ROWS[1]))
 
plt.scatter(X, Y, label='block', s=100, alpha=0.7)

plt.title('Graph multiplication one thread')
plt.xlabel('Size')
plt.ylabel('time(sec)')
plt.legend()
plt.grid()
plt.show()
```
